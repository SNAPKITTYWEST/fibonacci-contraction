````python
# ┌─────────────────────────────────────────────────────────────────┐
# │ CELL 2 — PHINARY REPRESENTATION (ZECKENDORF)                   │
# │ State handover: receives PHI, WORM_STATE from cell 1           │
# │ >++[->++<]>. (Brainfuck: Fibonacci tape advance)               │
# └─────────────────────────────────────────────────────────────────┘

def fib_sequence(n):
    a, b = 1, 1
    seq = [1, 1]
    while len(seq) < n:
        a, b = b, a + b
        seq.append(b)
    return seq

def is_prime(n):
    if n < 2: return False
    if n == 2: return True
    if n % 2 == 0: return False
    return all(n % i != 0 for i in range(3, int(n**0.5)+1, 2))

FIBS = fib_sequence(25)
FIB_PRIMES = [f for f in FIBS if is_prime(f)]

def zeckendorf(n):
    """Phinary (base-φ) representation via Zeckendorf decomposition.
    Returns list of (fibonacci_number, bit) pairs — no two consecutive 1s.
    """
    fibs = sorted([f for f in FIBS if f <= n], reverse=True)
    bits = []
    rem = n
    for f in fibs:
        if f <= rem:
            bits.append((f, 1))
            rem -= f
            # skip next (no consecutive 1s)
        else:
            bits.append((f, 0))
    return bits

def phinary_str(n):
    z = zeckendorf(n)
    return ''.join(str(b) for _, b in z)

# Display phinary table
print(f"{'N':>4} {'Phinary':>20} {'Zeckendorf sum':>35} {'Fib Prime?':>12}")
print('─' * 75)
for n in range(1, 21):
    z = zeckendorf(n)
    fib_sum = ' + '.join(f'F({FIBS.index(f)+1})={f}' for f,b in z if b)
    ph = phinary_str(n)
    prime_mark = '★ PRIME' if is_prime(n) and n in FIB_PRIMES else ''
    print(f"{n:>4} {ph:>20} {fib_sum:>35} {prime_mark:>12}")

print(f"\nFibonacci primes: {FIB_PRIMES}")

# Seal + hand off state
c2 = worm_seal(2, {'fibs': FIBS[:10], 'fib_primes': FIB_PRIMES, 'prev_seal': c1['seal']})
print(f"\nWORM SEAL: {c2['seal']}")
````
